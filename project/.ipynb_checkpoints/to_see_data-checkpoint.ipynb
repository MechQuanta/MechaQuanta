```python
import scipy.io

# List of your .mat files
file_names = ["2DIBPRegion1.mat", "2DIBPRegion2.mat", "2DIBPRegion3.mat", "2DIBPRegion4.mat"]

for file_name in file_names:
    try:
        print(f"--- Contents of {file_name} ---")
        mat_data = scipy.io.loadmat(file_name)

        # The data is loaded as a dictionary.
        # We can iterate through the items to see the variable names and their content.
        for key, value in mat_data.items():
            # We don't need to print the header, version, and globals information
            if not key.startswith('__'):
                print(f"Variable: '{key}'")
                print("Data:")
                print(value)
        print(len(file_name) + 16)
    except FileNotFoundError:
        print(f"Error: The file '{file_name}' was not found in the same directory as the script.")
    except Exception as e:
        print(f"An error occurred while reading {file_name}: {e}")
```
